```python
import numpy as np
import pandas as pd
import seaborn
from matplotlib import pyplot as plt
```

```python
xmlrpc_local = pd.read_csv('data/xmlrpc_100it_local.csv', sep=' ', header=None, names=['function_name', 'duration'])
xmlrpc_network = pd.read_csv('data/xmlrpc_100it_network.csv', sep=' ', header=None, names=['function_name', 'duration'])
grpc_local = pd.read_csv('data/grpc_100it_local.csv', sep=' ', header=None, names=['function_name' ,'duration'])
grpc_network = pd.read_csv('data/grpc_100it_network.csv', sep=' ', header=None, names=['function_name', 'duration'])
grpc_local[:17]

xmlrpc_local['duration'] *= 1000
xmlrpc_local['duration'] *= 1000
xmlrpc_network['duration'] *= 1000
grpc_local['duration'] *= 1000
grpc_network['duration'] *= 1000
```

```python
print('xmlrpc_local')
print(xmlrpc_local.describe())
print('xmlrpc_network')
print(xmlrpc_network.describe())
```

```python

fig, ax = pyplot.subplots(figsize=(15, 15))

g = seaborn.boxplot(x="function_name", y="duration", data=xmlrpc_local, color='blue')
fig.add_subplot(g)
g.set_title('XMLRPC - local vs rede')

g = seaborn.boxplot(x="function_name", y="duration", data=xmlrpc_network, color='red')
fig.add_subplot(g)

plt.xticks(rotation=90)

print('local')
print(xmlrpc_local.describe())
print('network')
print(xmlrpc_network.describe())
```

```python
fig, ax = pyplot.subplots(figsize=(15, 15))

g = seaborn.boxplot(x="function_name", y="duration", data=grpc_local, color='blue')
fig.add_subplot(g)
g.set_title('GRPC - local vs rede')


g = seaborn.boxplot(x="function_name", y="duration", data=grpc_network, color='red')
fig.add_subplot(g)

plt.xticks(rotation=90)
print('local')
print(grpc_local.describe())
print('network')
print(grpc_network.describe())

```
